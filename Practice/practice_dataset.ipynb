```python
city_temp = dt.groupby('City')['Temperature_Max (°C)'].mean()
plt.plot(city_temp.index, city_temp.values)
plt.xlabel('City')
plt.ylabel('Avg Max Temp (°C)')
plt.xticks(rotation=45)
```

```python
dt.sort_values(
    by=['Wind_Speed (km/h)', 'Rainfall (mm)'], 
    ascending=[False, True]
    )[['City','Wind_Speed (km/h)', 'Rainfall (mm)']].head()
```

```python
# filter State with less temperature , less rainfall chance , less humidity , with normal pressure and windspeed
filtered_dt = dt[
    (dt['Temperature_Min (°C)'].between(15, 25) ) & 
    (dt['Rainfall (mm)'] <=5) &
    (dt['Wind_Speed (km/h)'].between(3,18)) &
    (dt['Humidity (%)'] <=33) &
    (dt['Pressure (hPa)'].between(1005, 1010)) 
][['State', 'Temperature_Min (°C)','Humidity (%)','Rainfall (mm)', 'Wind_Speed (km/h)', 'Pressure (hPa)']]

filtered_dt.head()
```

```python
dt.groupby('Date')['Temperature_Avg (°C)'].mean()
```

```python
# Monthly average temperature

dt['Date'] = pd.to_datetime(dt['Date'])
dt['Month'] = dt['Date'].dt.month
dt['Day'] = dt['Date'].dt.day

dt.groupby('Month')['Temperature_Avg (°C)'].mean().head()
```

```python
# catogerizing temperature with label cool , warm , extreme

dt['Temp_level'] = pd.cut(
    dt['Temperature_Max (°C)'],
    bins=[18, 30, 35,45],
    labels=['cool', 'warm', 'extreme']
)
```

```python
month_temp_level = dt.groupby(['Month', 'Temp_level', 'State']).size()
month_temp_level
```

```python
# Filtering & Conditional Selection

dt[(dt['Temperature_Max (°C)'] > 40) & (dt['AQI'] > 200)].sort_values(by='Rainfall (mm)', ascending=False).head(3)
```
