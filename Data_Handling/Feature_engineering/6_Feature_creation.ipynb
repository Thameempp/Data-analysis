### What it is
- Creating new features from existing ones using logic or domain knowledge.

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "height": [170, 160, 180],
    "weight": [68, 55, 75]
})

# creating new feature (bmi) from features weight and height
df["BMI"] = df["weight"] / (df["height"] / 100) ** 2
df
```

```python
df = pd.DataFrame({
    "price": [100, 200, 150],
    "quantity": [2, 1, 4]
})

# creating new feature (total price) from price and quantity
df["total_price"] = df["price"] * df["quantity"]
df
```

```python
df["is_high_price"] = (df["price"] > 150).astype(int)
```
