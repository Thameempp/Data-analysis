### what is mean by imbalanced dataset:
- An imbalanced dataset is a dataset where the number of samples in one class is much higher than the number of samples in another class.

### Example :
#### Fraud detection
- Fraud cases: 2%
- Non-fraud cases: 98%
- â†’ This is an imbalanced dataset


### quick check :
- If a model predicts only the majority class and still gets high accuracy, your dataset is imbalanced.

### why it is a problem:
- Models become biased toward the majority class
- High accuracy but poor real performance
- Minority class gets ignored
