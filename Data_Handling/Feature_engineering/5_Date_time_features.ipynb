### Date & Time Feature Engineering
#### what is it :
- Extract meaningful numerical features from date/time columns.
#### why needed :
- Models donâ€™t understand timestamps directly.

### common date features
- Year, Month, Day
- Day of week
- Is weekend
- Hour, Minute
- Time difference

```python
import pandas as pd

df = pd.DataFrame({
    "order_date": ["2024-01-05", "2024-01-06", "2024-01-07"]
})

df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
df
```

```python
df["year"] = df["order_date"].dt.year
df["month"] = df["order_date"].dt.month
df["day"] = df["order_date"].dt.day
df["day_of_week"] = df["order_date"].dt.dayofweek
df["is_weekend"] = df["day_of_week"].isin([5, 6]).astype(int)

df
```

```python
df["days_since_order"] = (pd.Timestamp.today() - df["order_date"]).dt.days
```
