```python
# what is encoding : 
# process of converting categorical data into numeric form so that machine learning algorithm can process it


# types of categorical data:
# nomial - categorical with no order.   example: color, city, gender
# ordinal - categorical with meaningful order. example : low < medium < high
```

```python
# use in ml :
# Data encoding enables machine learning models to understand, process, and learn from categorical data correctly and efficiently.
```

## Label encoding

- It assigns an integer (0, 1, 2, …) to each unique category.
- Label encoding is based on alphebetical order , first  0 , then 1 , .....

```python

import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
df = pd.DataFrame({
    'color':('red','green','blue','red','blue','blue','green')
})
df
```

```python
encoded_colors = encoder.fit_transform(df)
encoded_colors
```

```python
y = ["Pass", "Fail", "Pass", "Fail", "Pass"]
```

```python
y_encoded = encoder.fit_transform(y)

print(y_encoded)
print(encoder.classes_)
```

```python
# fit

y1 = ["Yes", "No", "Yes", "No"]
k1 = encoder.fit(y1)
# it internally learn, but does not change the data
```

```python
# transform
y1_encoded = k1.transform(y)
print(y1_encoded)
# uses already-learned information to convert data.
# here k1 is  data that already internally learned using fit(y)
```

```python
# fit_transform
y1 = ["Yes", "No", "Yes", "No"]
encoder.fit_transform(y1)
# it changes the data
# fit_transform(y) = fit(y) + transform(y)
```

fit() learns parameters from data, transform() applies them, and fit_transform() performs both in one step.

### Ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd
```

```python
df = pd.DataFrame({
    'size':['small', 'large', 'medium', 'large', 'medium', 'small']
})
```

```python

enc = OrdinalEncoder(categories=[['small', 'medium', 'large']])
```

```python
enc.fit_transform(df[["size"]])
```

```python
enc.transform([['small']])
```

```python
enc.transform([['large']])
```

### One hot encoding

```python
# converts a categorical feature into multiple binary(0/1) columns, one column per catefory

# Only one column is “hot” (1) for each row, others are 0.

# Why One-Hot Encoding is Needed: Label Encoding creates fake order for nominal data.
```

```python
'''
Red → 0
Blue → 1
Green → 2
'''
# and it may think 
# Green > Blue > Red 
# so that why we use onehot encoding
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
encode = OneHotEncoder()
```

```python
color = ["Red", "Blue", "Green", "Red"] # (4,)
# What OneHotEncoder Expects : (number_of_samples, number_of_features)
# that means : (rows, columns) , for us : (4 sample,  1 color)
# it expects (4,1) 
colors = np.array(["Red", "Blue", "Green", "Red"]).reshape(-1,1)
# reshape(-1,1) : here -1 means : Automatically calculate this dimension for me , so 4 row 
# so this transforms to like this :
# reshape(-1, 1) = reshape(4, 1)
'''
[['Red']
 ['Blue']
 ['Green']
 ['Red']]
 '''
```

```python
ohe = OneHotEncoder(sparse_output=False)
```

```python
encoded_colors = ohe.fit_transform(colors)

print(encoded_colors)
```
