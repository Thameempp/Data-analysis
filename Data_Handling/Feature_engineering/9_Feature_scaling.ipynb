### Feature Scaling
#### What it is :
- Scaling ensures all features are on similar numerical ranges.

#### Why needed

##### Important for:

- Logistic Regression
- KNN
- SVM
- Neural Networks

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "income": [20000, 30000, 500000, 1000000]
})
```

```python
# StandardScaler (Z-score)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[["income"]])
df_scaled
```

```python
# MinMaxScaler (0â€“1)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df[["income"]])
df_scaled
```

```python
# RobustScaler (handles outliers)

from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
df_scaled = scaler.fit_transform(df[["income"]])
df_scaled
```
