Key differences

Syntax: callable vs direct expressions/literals.
Evaluation context: callables receive the current DataFrame (handy in method chains); direct expressions reference variables in scope.
Behavior: result is the same when the callable returns the same indexers, but callables improve readability and composability in chained pipelines.


Summary: Use the callable form when you want selection logic to be evaluated with the DataFrame passed into .loc (especially in chained operations); use direct boolean/list form for straightforward, immediate selection.

```python
# Merge


sf1 = pd.DataFrame({
    'label':['a','b','c','d'],
    'age': [10,20,30,40],
    'name':['john','jacob','sara','ali']
})

sf2 = pd.DataFrame({
    'label':['a','b','c'],
    'name':['book','pen','lap']
})

pd.merge(sf1, sf2, on='label') # default it will be inner join
```

```python
# types , using 'how'
'''
| how     | Meaning          |
| ------- | ---------------- |
| `inner` | common keys only |
| `left`  | all from left    |
| `right` | all from right   |
| `outer` | all keys         |
'''
```

```python
pd.merge(sf1, sf2, on='label', how='left')
```

```python
# merge with different column name 

pd.merge(sf1, sf2, left_on='label', right_on='label')
```

```python
# Join
```

```python
# concatenation vs merge

'''
| Task          | Use              |
| ------------- | ---------------- |
| Stack rows    | `concat(axis=0)` |
| Add columns   | `concat(axis=1)` |
| Match by key  | `merge()`        |
| Join on index | `join()`         |
'''
```
