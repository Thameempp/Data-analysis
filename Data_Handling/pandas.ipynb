### Updating & Replacing

```python
# updating single cell
df.loc[1] # students Liam
df.loc[1, 'City'] = 'Pune'
af = df.rename(columns={'Attendance (%)':'Attendance'})
af
```

```python
# using iloc
df.iloc[3, 3]='CSE' # changed branch of student Noah with index 3 from ECE to CSE
df
```

```python
# Updating Multiple Column at once
df['Score'] = df['Score'] + 1
df
```

```python
# Add a column based on another column

df['Result'] = df['Score'].apply(lambda x: 'Pass' if x>= 70 else 'Fail')  
df
```

### Deleting & Cleaning & Null Handling

```python
# Deleting a specific content
df = df.drop(9, axis=0)
```

```python
df.loc[df['Name'] =='Emma', 'Score'] = np.nan
df.loc[df['Student_ID'] ==108, 'Attendance'] = np.nan
df.loc[df['Student_ID'] ==103, 'Branch'] = np.nan
```

```python
df
```

```python
# Filling NaN Values with zero clean the data
df['Score'] =df['Score'].fillna(0)
df['Score']

df.loc[df['Student_ID']==103,'Branch']='ME'
df
```
