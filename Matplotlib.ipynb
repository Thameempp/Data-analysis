#   Line Plot 

```python
#  line styles : linestyle or 'colorTYPE' : linestyle='--'  , if color is red , then : 'r--'
#  types ;     '-', '--', ':', '-.'
```

```python
# subplot :  row * column

# plt.subplot(2,2,n). -> 2×2 Grid (Most Common)
# plt.subplot(3,3,n). -> 3×3 Grid 
# plt.subplot(1,4,n). -> 1×4 (Horizontal Layout)
# plt.subplot(2,2,n). -> 4×1 (Vertical Layout)
# plt.subplot(2,2,n). -> 2×2 Grid (Most Common)

# fig → the overall figure object (the entire canvas)
# axes → a 2D array of Axes objects, shaped (2, 2)  :
    
# axes[0,0] → top-left
# axes[0,1] → top-right
# axes[1,0] → bottom-left
# axes[1,1] → bottom-right

```

```python
# sample data 
p = [2,4,6,8,9]
q = [4,16,36,64,81]              # here p and q must have same dimension   

# creating a line plot
plt.plot(p,q)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# creating a customized line plot

plt.plot(p,q, color='blue', linestyle='--', marker='o', linewidth=1, markersize=9)
```

```python
# multiple plots 

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]   # y = x^2
y2 = [1, 2, 3, 4, 5]     # y = x

plt.figure(figsize=(8,4))

plt.subplot(2,2,1)
plt.plot(x, y1, color='green')
plt.title("plot 1")            # parabolic

plt.subplot(2,2,2)
plt.plot(y1, x, color='red')
plt.title("plot 2")            # inverse of standard parabola. (horizontal).  x = y^2

plt.subplot(2,2,3)
plt.plot(x, y2, color='blue')
plt.title('plot 3')                 

plt.subplot(2,2,4)
plt.plot(y2, x, color='orange')
plt.title('plot 4')
```

```python
plt.figure(figsize=(4,3))
plt.plot(x, y1, color='green', label='y = x^2', linestyle='--')
plt.plot(x, y2, color='red', label='y = x', linewidth=2)
plt.title('multiple line one plot')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y1, 'r--', label='-- dashed')
plt.plot(x, y2, 'b:', label=': dotted')
# plt.plot(x, y3, 'g-.', label='-. dashdot')

plt.legend()
plt.title("Different Line Styles")
plt.grid(True)
plt.show()
```

```python
# Line Transparency  example;    alpha= 0.1 (more trans..) ,  0.8 (less trans..)
```

```python
plt.figure(figsize=(4,3))
plt.plot(x, y1, color='red', alpha=0.2, linewidth=8, label='Transparent')
plt.plot(x, y2, color='blue', alpha=0.8, linewidth=4, label='Less Transparent')

plt.legend()
plt.title("Line Transparency Example")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.step(x, y1, 'r--', linewidth=3,)
plt.title('plot 1')
plt.xlabel('x')
plt.ylabel('y1')
plt.grid(True)


plt.subplot(1,3,2)
plt.step(x, y2, 'g:', linewidth=3)
plt.title('plot 2')
plt.xlabel('x')
plt.ylabel('y2')
plt.grid(True)


plt.subplot(1,3,3)
plt.step(y1, x, 'b-.', linewidth=3)
plt.title('plot 3')
plt.xlabel('y1')
plt.ylabel('x')
plt.grid(True)

plt.tight_layout()
plt.show()

```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

fig, axes = plt.subplots(2, 2, figsize=(8,6))


axes[0,0].plot(x, y, color='red')
axes[0,0].set_title("Plot 1: y = x²")  # top left

axes[0,1].plot(y, x, color='blue')
axes[0,1].set_title("Plot 2: inverse parabola")  # top right

axes[1,0].plot(x, [i*2 for i in y], color='green')     # [i*2 for i in y] -> multiply each element in y with 2 , so y = 2 * x^2
axes[1,0].set_title("Plot 3: y = 2x² (scaled up)")  

axes[1,1].plot(y, [i/2 for i in x], color='purple')
axes[1,1].set_title("Plot 4: flattened inverse") 

plt.tight_layout()
plt.show()
```

```python
plt.plot(x, y1, color='green')
plt.fill_between(x, y1, color='lightgreen', alpha=0.4)
plt.title("Line Plot with Filled Area")
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y1, color='blue', marker='o')
plt.title("Annotated")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add a note at the max point
plt.annotate('Peak', xy=(3, 9), xytext=(2, 15),         # xytext=(2, 15) -> position of annoted word
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate('Minimum', xy=(1, 1), xytext=(1, 12),
             arrowprops=dict(facecolor='orange', arrowstyle='->'))

plt.grid(True)
plt.show()
```
