#   Line Plot 

```python
#  line styles : linestyle or 'colorTYPE' : linestyle='--'  , if color is red , then : 'r--'
#  types ;     '-', '--', ':', '-.'
```

```python
# subplot :  row * column

# plt.subplot(2,2,n). -> 2×2 Grid (Most Common)
# plt.subplot(3,3,n). -> 3×3 Grid 
# plt.subplot(1,4,n). -> 1×4 (Horizontal Layout)
# plt.subplot(2,2,n). -> 4×1 (Vertical Layout)
# plt.subplot(2,2,n). -> 2×2 Grid (Most Common)

# fig → the overall figure object (the entire canvas)
# axes → a 2D array of Axes objects, shaped (2, 2)  :
    
# axes[0,0] → top-left
# axes[0,1] → top-right
# axes[1,0] → bottom-left
# axes[1,1] → bottom-right

```

```python
# sample data 
p = [2,4,6,8,9]
q = [4,16,36,64,81]              # here p and q must have same dimension   

# creating a line plot
plt.plot(p,q)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# creating a customized line plot

plt.plot(p,q, color='blue', linestyle='--', marker='o', linewidth=1, markersize=9)
```

```python
# multiple plots 

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]   # y = x^2
y2 = [1, 2, 3, 4, 5]     # y = x

plt.figure(figsize=(8,4))

plt.subplot(2,2,1)
plt.plot(x, y1, color='green')
plt.title("plot 1")            # parabolic

plt.subplot(2,2,2)
plt.plot(y1, x, color='red')
plt.title("plot 2")            # inverse of standard parabola. (horizontal).  x = y^2

plt.subplot(2,2,3)
plt.plot(x, y2, color='blue')
plt.title('plot 3')                 

plt.subplot(2,2,4)
plt.plot(y2, x, color='orange')
plt.title('plot 4')
```

```python
plt.figure(figsize=(4,3))
plt.plot(x, y1, color='green', label='y = x^2', linestyle='--')
plt.plot(x, y2, color='red', label='y = x', linewidth=2)
plt.title('multiple line one plot')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y1, 'r--', label='-- dashed')
plt.plot(x, y2, 'b:', label=': dotted')
# plt.plot(x, y3, 'g-.', label='-. dashdot')

plt.legend()
plt.title("Different Line Styles")
plt.grid(True)
plt.show()
```

```python
# Line Transparency  example;    alpha= 0.1 (more trans..) ,  0.8 (less trans..)
```

```python
plt.figure(figsize=(4,3))
plt.plot(x, y1, color='red', alpha=0.2, linewidth=8, label='Transparent')
plt.plot(x, y2, color='blue', alpha=0.8, linewidth=4, label='Less Transparent')

plt.legend()
plt.title("Line Transparency Example")
plt.grid(True)
plt.show()
```

## Step plot

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.step(x, y1, 'r--', linewidth=3)
plt.title('plot 1')
plt.xlabel('x')
plt.ylabel('y1')
plt.grid(True)


plt.subplot(1,3,2)
plt.step(x, y2, 'g:', linewidth=3)
plt.title('plot 2')
plt.xlabel('x')
plt.ylabel('y2')
plt.grid(True)


plt.subplot(1,3,3)
plt.step(y1, x, 'b-.', linewidth=3)
plt.title('plot 3')
plt.xlabel('y1')
plt.ylabel('x')
plt.grid(True)

plt.tight_layout()
plt.show()

```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

fig, axes = plt.subplots(2, 2, figsize=(8,6))


axes[0,0].plot(x, y, color='red')
axes[0,0].set_title("Plot 1: y = x²")  # top left

axes[0,1].plot(y, x, color='blue')
axes[0,1].set_title("Plot 2: inverse parabola")  # top right

axes[1,0].plot(x, [i*2 for i in y], color='green')     # [i*2 for i in y] -> multiply each element in y with 2 , so y = 2 * x^2
axes[1,0].set_title("Plot 3: y = 2x² (scaled up)")  

axes[1,1].plot(y, [i/2 for i in x], color='purple')
axes[1,1].set_title("Plot 4: flattened inverse") 

plt.tight_layout()
plt.show()
```

#### Filled area line plot

```python
plt.plot(x, y1, color='green')
plt.fill_between(x, y1, color='lightgreen', alpha=0.4)
plt.title("Line Plot with Filled Area")
plt.grid(True)
plt.show()
```

#### Line plot with Annotation

```python
plt.plot(x, y1, color='blue', marker='o')
plt.title("Annotated")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add a note at the max point
plt.annotate('Peak', xy=(3, 9), xytext=(2, 15),         # xytext=(2, 15) -> position of annoted word
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate('Minimum', xy=(1, 1), xytext=(1, 12),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.grid(True)
plt.show()
```

| Part                | Explanation                                       |
| ------------------- | ------------------------------------------------- |
| `'Peak'`            | Text that will appear on the chart                |
| `xy=(3, 9)`         | **Exact data point** being highlighted (x=3, y=9) |
| `xytext=(2, 15)`    | **Position where the text is displayed**          |
| `arrowprops`        | Controls arrow appearance                         |
| `facecolor='black'` | Arrow color                                       |
| `arrowstyle='->'`   | Arrow shape                                       |


## Bar plot

```python
# plt.bar(x, height)
```

```python
# vertical bar plot
categories = ['A', 'B', 'C']
values = [10, 15, 7]

plt.bar(categories, values)
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Simple bar plot')
plt.show()
```

```python
# horizontal bar plot
plt.barh(categories, values)
```

```python
# Grouped bar plot
# example 1
x = np.arange(len(categories))
width = 0.35
plt.bar(x-width/2, [10, 15, 7], width, label='2024')
plt.bar(x+width/2, [12, 18, 9],width, label='2025', color='black')
```

```python
# explanation of grouped bar plot
"""
x + width/2 → positions of the bars on the X-axis
[12, 18, 9] → Y-axis values (bar heights)
width → thickness of each bar
label='2025' → legend name
color='black' → bar color
"""

'''
x positions →   Jan     Feb     Mar
2024 bars  →   10      15      7
2025 bars  →   12      18      9
'''
```

```python
# example 2
products = ['Laptops','Mobile','Tablet']
sales_2023 = [45, 65, 30]
sales_2024 = [70, 95, 25]
# x position
x = np.arange(len(products))

# bar width
width = 0.35

# plotting grouped bars
plt.bar(x-width/2, sales_2023, width, label='2023', color='green')
plt.bar(x+width/2, sales_2024, width, label='2024',color='brown')

# label
plt.xlabel('products')
plt.ylabel('sales (in units)')
plt.title('Product sales comparison')
plt.xticks(x, products)
plt.legend()
plt.show()
```

## Histogram

```python

data = [5, 19, 35, 60, 65, 81, 75, 80, 85, 90]

plt.hist(data)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Basic Histogram")
plt.show()
```

```python
plt.hist(data, bins=5, edgecolor='black') # bins=5 → divides data into 5 ranges
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram with Custom Bins")
plt.show()
```

#### Multiple Histograms (Comparison)

```python
marks_2024 = [50, 55, 60, 65, 70, 75, 80]
marks_2025 = [55, 60, 65, 70, 75, 85, 90]

plt.hist(marks_2024, bins=5, alpha=0.6, label='2024')
plt.hist(marks_2025, bins=5, alpha=0.6, label='2025')

plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Marks Distribution Comparison")
plt.legend()
plt.show()
```

Advanced Histogram (Density + Range Control)

```python
data = np.random.normal(70, 10, 1000)  # mean=70, std=10 # It generates 1000 random numbers that follow a normal (bell-shaped) distribution

# np.random.normal() → realistic data # np.random.normal() - Generates random values from a normal distribution
# Also called Gaussian distribution
# density=True → area under histogram = 1

plt.hist(data, bins=30, density=True, edgecolor='black')
plt.xlabel("Values")
plt.ylabel("Density")
plt.title("Advanced Histogram (Density)")
plt.show()
```

## Scatter plot

```python
x = [1, 2, 3, 4, 5, 9, 7, 8]
y = [3, 4, 5, 16, 17, 18, 49, 64]

plt.scatter(x, y)

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Basic Scatter Plot")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 8, 20, 12]
sizes = [100, 200, 150, 300, 180]


plt.scatter(x, y, s=sizes, alpha=0.6)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter with Size")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

colors = np.random.rand(50)

plt.scatter(x, y, c=colors, alpha=0.7)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Advanced Scatter Plot")
plt.show()
```

## Box plot

```python
data = [45, 47, 49, 50, 55, 60, 65, 70, 75, 100]

plt.boxplot(data)
plt.title("Basic Box Plot")
plt.ylabel("Values")
plt.show()
```

```python
# Box plot with multiple group
marks_A = [50, 55, 60, 65, 70]
marks_B = [45, 50, 55, 60, 90]

plt.boxplot([marks_A, marks_B], labels=['Class A', 'Class B'])
plt.title("Marks Comparison")
plt.ylabel("Marks")
plt.show()
```

```python
# Horizontal box plot 
data = [45, 50, 55, 60, 65, 70, 75, 100]

plt.boxplot(data, vert=False)
plt.title("Horizontal Box Plot")
plt.xlabel("Values")
plt.show()
```

```python
data = np.random.normal(70, 10, 100)

plt.boxplot(
    data,
    patch_artist=True,
    showfliers=True
)

plt.title("Advanced Box Plot")
plt.ylabel("Values")
plt.show()
```

## Pie chart

```python
sizes = [40, 30, 20, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels)
plt.title("Basic Pie Chart")
plt.show()
```

```python
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%', # shows percentage value on each slice # %1.1f -> one decimal place # %% -> display % symbol
    startangle=90 # not mendatory
)
plt.title("Pie Chart with Percentages")
plt.show()
```

```python
# explode
explode = [0.1, 0, 0, 0]  # highlight first slice

plt.pie(
    sizes,
    labels=labels,
    explode=explode,
    autopct='%1.1f%%',
    shadow=True
)
plt.title("Exploded Pie Chart")
plt.show()
```

```python
expl = [0.1, 0.3, 0.5, 0.2]
plt.pie(
    sizes,
    labels=labels,
    explode=expl,
    autopct='%1.0f%%',
    shadow=True
)
plt.title('practice explode')
plt.show()
```

```python
# example
companies = ['Apple', 'Samsung', 'Xiaomi', 'Others']
market_share = [35, 30, 20, 15]

plt.pie(
    market_share,
    labels=companies,
    autopct='%1.0f%%',
    startangle=140
)
plt.title("Smartphone Market Share")
plt.show()
```

## Area Plot / stack plot

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 2, 6, 4]

plt.fill_between(x, y)
plt.title("Simple Area Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

### Stacked area plot

```python
x = [2019, 2020, 2021, 2022, 2023]

product_A = [20, 25, 30, 35, 40]
product_B = [15, 20, 25, 30, 35]
product_C = [10, 15, 20, 25, 30]

plt.stackplot(
    x,
    product_A,
    product_B,
    product_C,
    labels=['A', 'B', 'C']
)

plt.legend(loc='upper left')
plt.title("Stacked Area Plot")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```
