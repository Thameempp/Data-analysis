```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
print(df.head())
```

```python
# Basic Scatter plot
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
# Line plot

sns.lineplot(x='total_bill', y='tip', data=df) 
plt.show()
```

```python
# Bar plot

sns.barplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
# Count plot

sns.countplot(x='day', data=df)
plt.show()
```

```python
# Histogram

sns.histplot(df['total_bill'], bins=20, kde=True)
plt.show()
```

```python
sns.histplot(df['tip'], bins=20, kde=True)
plt.show()
```

## Pair plot

```python
sns.pairplot(df, hue='day')
```

## Join plot

```python
sns.jointplot(data=df, x='day', y='total_bill')
```

## violin plot

```python
sns.violinplot(data=df, x='size', y='total_bill')
```

## Heatmap

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap='coolwarm') # common cmap(color map) : coolwarm  correlations, viridis  general numeric, magma, plasma, inferno

# reverse cmap: 'coolwarm_r'
```

```python
sns.heatmap(corr, annot=True)
```
