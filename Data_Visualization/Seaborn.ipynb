```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

## Relational plot

```python
# Basic Scatter plot
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
# Line plot

sns.lineplot(x='total_bill', y='tip', data=df) 
plt.show()
```

```python
# Bar plot

sns.barplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
# Count plot

sns.countplot(x='day', data=df)
plt.show()
```

## Distribution plot

```python
# Histogram

sns.histplot(df['total_bill'], bins=20, kde=True)
plt.show()
```

```python
sns.histplot(df['tip'], bins=20, kde=True)
plt.show()
```

```python
# displot

sns.displot(data=df, x='total_bill', kind='hist', bins=3)
```

```python
sns.displot(data=df, x='day', kind='hist')
```

```python
# hue parameter
sns.displot(data=df, x='tip', kind='hist', hue='sex')
```

```python
# element -> step
sns.displot(data=df, x='tip', kind='hist', hue='sex', element='step')
```

```python
kf = sns.load_dataset('diamonds')
kf
```

```python
sns.displot(data=kf, x='depth', kind='hist', element='step', hue='clarity')
```

## Matrix plot

```python
# Heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap='coolwarm') # common cmap(color map) : coolwarm  correlations, viridis  general numeric, magma, plasma, inferno

# reverse cmap: 'coolwarm_r'
```

```python
sns.heatmap(corr, annot=True)
```
